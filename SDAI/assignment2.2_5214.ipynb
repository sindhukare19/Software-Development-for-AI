```python
import matplotlib.pyplot as plt

animals = ['Penguins', 'Polar bears', 'Lions', 'Otters']
numbers = [30, 40, 20, 10]
colors = ['#FFD700', '#ADD8E6', '#FFC0CB', '#9ACD32']

plt.pie(numbers, labels=animals, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.show()
```

```python
#import matplotlib.pyplot as plt

movies = ['Comedy', 'Action', 'Romance', 'Drama', 'Sci-Fi']
likes = [10, 8, 15, 2, 5]

plt.barh(movies, likes)
plt.xlabel('Number of people')
plt.ylabel('Movie type')
plt.show()
```

```python
import matplotlib.pyplot as plt
#importing numpy
import numpy as np

m = ['April', 'May', 'June', 'July']
c = [780, 1050, 3056, 5025]
a = [315, 400, 1000, 1500]

bar_width = 0.35
index = np.arange(len(m))

plt.bar(index, c, bar_width, label='C')
plt.bar(index + bar_width, a, bar_width, label='A')

plt.xlabel('M')
plt.ylabel('Number of visitors')
plt.xticks(index + bar_width / 2, m)
plt.legend()
plt.show()
```
